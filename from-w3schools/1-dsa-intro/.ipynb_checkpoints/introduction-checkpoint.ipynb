# What are Data Structures?
Data Structures are a way to store a data. Data structures are meant to give us the possibility to manage large amount of data eficiently! Often Data Structures are used by an Algorithm to help in managing and organizing data, reduce space and time complexity, and increase effiency both from the Data Strucutres and the Algorithm it self.

**In Computer Science there are two fundamentals Data Structures:**
- Primitive Data Structures (in Python): Primitive data structures are basic data structures that is provided by the programming language it self, such as fundamental data types like: int, str, float, char, boolean.

- Abstract Data Structures (in Python): Abstract data strucutres are higher-level data structures that are built using primitive data types and provide more complex and specialized operations. Some common examples of abstract data structures are array, linked-lists, stacks, queues, trees, and graphs.
